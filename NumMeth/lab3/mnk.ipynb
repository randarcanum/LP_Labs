```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y_x(x, coef):
    return np.sum([x**(coef.size-i-1)*a for i, a in enumerate(coef)])
```

```python
f = open("x_y.txt", "r")
x0 = np.array(f.readline().split()).astype(np.double)
y0 = np.array(f.readline().split()).astype(np.double)
f.close()
```

```python
x1 = np.linspace(0, 4.5, 50)
f = open("func.txt", "r")
coef = np.array(f.readline().split()).astype(np.double)
y1 = np.array([y_x(x, coef) for x in x1])
coef = np.array(f.readline().split()).astype(np.double)
y2 = np.array([y_x(x, coef) for x in x1])
f.close()
```

```python
plt.xticks([x/2 for x in range(10)])
plt.yticks([x-1 for x in range(9)])
plt.xlim(0, 4.5)
plt.ylim(-1.0, 7.0)
y0_line = plt.scatter(x0, y0, label="Исходные значения")
y1_line = plt.plot(x1, y1, label="y = 1.53x - 1.1")
y2_line = plt.plot(x1, y2, label="y = 0.371x^2 - 0.138x + 0.293")
plt.grid(color='r', linestyle='-', linewidth=0.5)
plt.legend()
plt.show()
```
